```python
# Convert Order Date and Ship Date into datetime format

clean_df['Order Date'] = pd.to_datetime(clean_df['Order Date'])

clean_df['Ship Date'] = pd.to_datetime(clean_df['Ship Date'])

print("Date conversion completed successfully.")
```

```python
# Reset the DataFrame index
clean_df = clean_df.reset_index(drop=True)

# Create new Row ID starting from 1
clean_df['Row ID'] = range(1, len(clean_df) + 1)

# Make Row ID the first column
cols = ['Row ID'] + [col for col in clean_df.columns if col != 'Row ID']
clean_df = clean_df[cols]

# Verify
clean_df.head()
```

 Create Order Year

The year of each order helps us compare business performance across different years.

```python
clean_df['Order Year'] = clean_df['Order Date'].dt.year

clean_df[['Order Date', 'Order Year']].head()
```

```python
# Step 10 – Create Order Month
##The month helps us analyze seasonal trends and monthly business performance.
clean_df['Order Month'] = clean_df['Order Date'].dt.month_name()

clean_df[['Order Date', 'Order Month']].head()
```

```python
##many companies prepare quarterly financial reports and use them to evaluate business performance.
clean_df['Order Quarter'] = clean_df['Order Date'].dt.quarter

clean_df[['Order Date', 'Order Quarter']].head()
```

```python
# Step 12 – Create Order Day Name

##Knowing the day of the week helps identify customer purchasing patterns.
clean_df['Order Day'] = clean_df['Order Date'].dt.day_name()

clean_df[['Order Date', 'Order Day']].head()
```

Calculate Shipping Days

Shipping Days represent the number of days taken to deliver an order.

Formula:

Shipping Days = Ship Date − Order Dat

```python
clean_df['Shipping Days'] = (
    clean_df['Ship Date'] - clean_df['Order Date']
).dt.days

clean_df[['Order Date', 'Ship Date', 'Shipping Days']].head()
```

Calculate Profit Margin

Profit Margin indicates how much profit is earned for every ₹100 (or $100) of sales.

Formula:

Profit Margin (%) = (Profit / Sales) × 100

```python
clean_df['Profit Margin (%)'] = (
    clean_df['Profit'] / clean_df['Sales']
) * 100

clean_df[['Sales', 'Profit', 'Profit Margin (%)']].head()
```

```python
new_columns = [
    'Order Year',
    'Order Month',
    'Order Quarter',
    'Order Day',
    'Shipping Days',
    'Profit Margin (%)'
]

print("New Features Created Successfully:\n")

for column in new_columns:
    print("✔", column)
```
